```python
import pandas as pd
```

```python
df = pd.read_excel('best-selling game consoles.xlsx',
                   sheet_name='consoles')
df.tail()
```

```python
companies = df.groupby('Company')[['Company']].count()
companies
```

```python
companies.columns = ['qtd']
companies = (companies.sort_values('qtd',ascending=False).reset_index())
companies
```

```python
market_value = pd.read_csv("company.txt")
market_value
```

```python
companies_mkt = companies.merge(market_value,on='Company',how='left')
companies_mkt
```

```python
soma = companies_mkt.sum()
soma
```

```python
print("Valor Médio: ", soma['Market Value']/soma['qtd'])
```
