```python
df = pd.read_excel('fastfood.xlsx')

df1 = pd.DataFrame({'produto':['relogio','anel','brinco','camisa'],

                    'loja':['Zimer','Piter','Piter','Loyyn'],

                    'valor':[100, 70, 20, 120]})

 

df2 = pd.DataFrame({'produto':['gravata','colar','bolsa'],

                    'loja':['Loyyn','Piter','Piter'],

                    'valor':[60, 120, 300]})

 

df3 = pd.DataFrame({'cidade':['Loyyn','Piter','Zimer'],

                    'loja':['New Romy','Piter','Golgo']})
```
